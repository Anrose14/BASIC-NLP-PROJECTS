```python
import pandas as pd
df=pd.read_csv('Data.csv',encoding='ISO-8859-1')
```

```python
df.head()
```

DIVIDING INTO TRAIN AND TEST

```python
train=df[df['Date']<'20150101']
test=df[df['Date']>'20150101']
```

```python
train.shape
```

```python
test.shape
```

TAKING THE TEXT INPUTS

```python
data=train.iloc[:,2:27]
data.head()
```

REPLACING ALL NON ALPHABETS TO SINGLE SPACES

```python
data.replace('[^a-zA-Z]',' ',inplace=True)
data.head()
```

```python
newindex=[i for i in range(25)]
data.columns=newindex
data.head()
```

LOWERCASING ALL TEXT DATA

```python
for index in newindex:
    data[index]=data[index].str.lower()
data.head()
```

```python
data.iloc[1,0:25]
```

```python
' '.join([str(i) for i in data.iloc[1,0:25]]) 
```

GETTING ALL COLUMN DATA TO A SINGLE TEXT DATA AND STORING IN A LIST

```python
headlines=[]

for row in range(len(data)):
    headlines.append(' '.join([str(i) for i in data.iloc[row,0:25]]))
```

```python
headlines[0:3]
```

VECTORIZATION

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.ensemble import RandomForestClassifier

c=CountVectorizer(ngram_range=(2,2))
trainx=c.fit_transform(headlines)
```

RANDOM FOREST CLASSIFIER MODEL

```python
rf=RandomForestClassifier(n_estimators=200,criterion='entropy')
rf.fit(trainx,train['Label'])
```

```python
test.head()
```

```python
headlinestest=[]

for row in range(len(test)):
    headlinestest.append(' '.join([str(i) for i in test.iloc[row,2:27]]))
```

```python
testx=c.transform(headlinestest)
predictions=rf.predict(testx)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
print(accuracy_score(test['Label'],predictions))
print(confusion_matrix(test['Label'],predictions))
```

```python
import matplotlib.pyplot as plt
import numpy as np

cm=confusion_matrix(test['Label'],predictions)
plt.imshow(cm,cmap=plt.cm.Blues)
plt.colorbar()
labels=['Positive','Negative']
tickmarks=np.arange(len(labels))

plt.xticks(tickmarks,labels)
plt.yticks(tickmarks,labels)

plt.xlabel('Actual')
plt.ylabel('Prediction')
```

```python
print(classification_report(test['Label'],predictions))
```

NAIVE BAYES MODEL

```python
from sklearn.naive_bayes import MultinomialNB
nb=MultinomialNB(alpha=1.0,class_prior=None,fit_prior=True)

nb.fit(trainx,train['Label'])
```

```python
predictionsnb=nb.predict(testx)
```

```python
print(accuracy_score(test['Label'],predictionsnb))
print(confusion_matrix(test['Label'],predictionsnb))
```

```python
cm1=confusion_matrix(test['Label'],predictionsnb)
plt.imshow(cm1,cmap=plt.cm.Blues)
plt.colorbar()
labels=['Positive','Negative']
tickmarks=np.arange(len(labels))

plt.xticks(tickmarks,labels)
plt.yticks(tickmarks,labels)

plt.xlabel('Actual')
plt.ylabel('Prediction')
```
