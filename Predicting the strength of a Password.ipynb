FUNCTION TO SPLIT THE CHARACTERS

```python
def worddividechar(inputs):
    character=[]
    for i in inputs:
        character.append(i)
    return character
worddividechar("kisb6")
```

VECTORIZATION

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tf=TfidfVectorizer(tokenizer=worddividechar)
```

```python
x=tf.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression(random_state=0,multi_class='multinomial')
```

```python
lr.fit(xtrain,ytrain)
```

```python
user=np.array(['bu6%&gbhAv'])
pred=tf.transform(user)
lr.predict(pred)
```

```python
predictions=lr.predict(xtest)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
cm=confusion_matrix(predictions,ytest)
print(cm)
print(accuracy_score(predictions,ytest))
```

```python
import matplotlib.pyplot as plt
plt.imshow(cm)
plt.colorbar()
labels=['0','1','2']
tickmarks=np.arange(len(labels))

plt.xticks(tickmarks,labels)
plt.yticks(tickmarks,labels)

plt.xlabel('Actual')
plt.ylabel('Prediction')
```

```python
print(classification_report(predictions,ytest))
```
