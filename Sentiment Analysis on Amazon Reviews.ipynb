HELPFULNESS COLUMN FROM NUMERATOR DENOMINATOR COLUMN

```python
import numpy as np
df['Helpfulness']=np.where(df['HelpfulnessDenominator']>0,df['HelpfulnessNumerator']/df['HelpfulnessDenominator'],-1)
```

CREATING UPVOTE COLUMN TO CATEGORIZE THE HELPFULNESS COLUMN

```python
df['%upvote']=pd.cut(df['Helpfulness'],bins=[-1,0,0.2,0.4,0.6,0.8,1],labels=['Empty','0-20%','20-40%','40-60%','60-80%','80-100%'])
```

REMOVING SCORE=3 ITEMS AS IT IS NEUTRAL NATURE

```python
df2=df[df['Score']!=3]
x=df2['Text']
```

CONVERTING SCORE CATEGORY TO 0 AND 1

```python
ydict={1:0,2:0,4:1,5:1}
y=df2['Score'].map(ydict)
```

TEXT VECTORIZATION USING COUNTVECTORIZER

```python
from sklearn.feature_extraction.text import CountVectorizer
c=CountVectorizer(stop_words='english')
x_c=c.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x_c,y)

from sklearn.linear_model import LogisticRegression
lr=LogisticRegression(max_iter=1000)

data=lr.fit(xtrain,ytrain)
```

FINDING IMPORTANCE OF INDIVIDUAL WORDS AND CLASSIFYING TO POSITIVE AND NEGATIVE

```python
w=c.get_feature_names_out()
coef=data.coef_.tolist()[0]
cf=pd.DataFrame({'Word':w,'Coefficient':coef})
cf.head()
cf.sort_values(['Coefficient','Word'],ascending=False)
```

AUTOMATING THE TRAINING USING FUNCTIONS

```python
def text_fit(x,y,nlp_model,ml_model,coef_show=1):
    x_c=nlp_model.fit_transform(x)
    
    print('features:{}'.format(x_c.shape[1]))
    
    xtrain,xtest,ytrain,ytest=train_test_split(x_c,y)
    
    data=ml_model.fit(xtrain,ytrain)
    
    accuracy=data.score(xtest,ytest)
    print('Accuracy: ',accuracy)
    
    if coef_show==1:
        
        w=c.get_feature_names_out()
        coef=data.coef_.tolist()[0]
        cf=pd.DataFrame({'Word':w,'Coefficient':coef})
        cf.sort_values(['Coefficient','Word'],ascending=False)
        
        print('Top 20 Positive words: ')
        print(cf.head(20))
        
        print('Top 20 Negative words: ')
        print(cf.tail(20))
```

```python
text_fit(x,y,c,lr)
```

AUTOMATING THE PREDICTING USING FUNCTIONS

```python
from sklearn.metrics import confusion_matrix,accuracy_score
def predict(x,y,nlp_model,ml_model):
    x_c=nlp_model.fit_transform(x)
    
    xtrain,xtest,ytrain,ytest=train_test_split(x_c,y)
    
    data=ml_model.fit(xtrain,ytrain)
    
    predictions=ml_model.predict(xtest)
    
    print(confusion_matrix(predictions,ytest))
    print(accuracy_score(predictions,ytest))
```

```python
predict(x,y,c,lr)
```

DUMMYCLASSIFIER FOR BASELINE COMPARISON

```python
from sklearn.dummy import DummyClassifier
text_fit(x,y,c,DummyClassifier(),0)
```

TEXT VECTORIZATION USING TFIDF VECTORIZER

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tf=TfidfVectorizer(stop_words='english')
text_fit(x,y,tf,lr,0)
```
